```python
import matplotlib.pyplot as plt
import seaborn as sns


num_vars = ['Area', 'Production', 'Annual_Rainfall', 'Fertilizer', 'Pesticide']

plt.figure(figsize=(14, 10))
for i, col in enumerate(num_vars, 1):
    plt.subplot(3, 2, i)
    sns.scatterplot(data= df, x=col, y='Yield', alpha=0.4)
    plt.title(f'{col} vs Yield')

plt.tight_layout()
plt.show()
```

```python
calculated_yield = df['Production'] / df['Area']
correlation = df['Yield'].corr(calculated_yield)
if correlation > 0.99:
    print("WARNING: Yield is calculated from Production and Area!")
    print("This creates data leakage in your model")
```
